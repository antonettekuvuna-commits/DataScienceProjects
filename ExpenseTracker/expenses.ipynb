# Expense Tracker

Managing personal finances requires tracking expenses.  
This project summarizes monthly spending and visualizes category breakdowns.


## Data Collection

We use a self-created dataset (`expenses.csv`) with Date, Category, and Amount.


```python
import pandas as pd
df = pd.read_csv("expenses.csv")
df.head()
```

## Data Cleaning

We convert dates into proper datetime format.  
This allows grouping by month for analysis.


```python
df['Date'] = pd.to_datetime(df['Date'])
```

## Analysis

We calculate monthly totals and visualize spending by category.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

monthly = df.groupby(df['Date'].dt.to_period("M"))['Amount'].sum()
print(monthly)

category = df.groupby("Category")['Amount'].sum()
category.plot(kind="bar")
plt.title("Expenses by Category")
plt.show()
```

## Conclusion

Food and Rent are the largest expenses.  
Visualizations help identify areas to cut costs and manage budgets better.
