# Nairobi Transport Data Analysis

Traffic congestion is a major issue in Nairobi.  
This project analyzes traffic volume by route and time to identify peak congestion periods.


## Data Collection

We use a synthetic dataset (`nairobi_traffic.csv`) with route, hour, and traffic volume.


```python
import pandas as pd
df = pd.read_csv("nairobi_traffic.csv")
df.head()
```

## Data Cleaning

We check for missing values and ensure data types are correct.  
This step ensures reliable analysis.


```python
df.info()
```

## Visualization

We plot traffic volume by hour and by route to identify peak congestion times.


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['hour'], bins=24)
plt.title("Traffic Volume by Hour")
plt.show()

sns.barplot(x="route", y="traffic_volume", data=df)
plt.title("Traffic Volume by Route")
plt.xticks(rotation=45)
plt.show()
```

## Conclusion

Morning (7–9 AM) and evening (5–7 PM) are peak congestion times.  
Westlands and Thika routes show the highest traffic volumes.  
This information can help city planners and transport companies optimize routes.
